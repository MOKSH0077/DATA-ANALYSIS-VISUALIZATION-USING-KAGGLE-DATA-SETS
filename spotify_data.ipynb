# SPOTIFY ,INDIAN YOUTH WITH PREMIUM SUBSCRIPTION
# SONGS PLAYED VS  LISTENING TIME

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spf=pd.read_csv("spotify_churn_dataset.csv")
print(spf)
```

```python
upd_spf=(spf['age']<25)&(spf['country']=='IN')&(spf['subscription_type']==('Premium'or'Student'))
spf_student_premium_india=spf[upd_spf]
print(spf_student_premium_india)
spf_student_premium_india.shape
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(12,10))


sns.barplot(x=spf_student_premium_india['songs_played_per_day'],y=spf_student_premium_india['listening_time'],hue=spf_student_premium_india['gender'])
plt.title("SPOTIFY SONGS PLAYED VS LISTENING TIME BY INDIAN YOUTH (PREMIUM)")
plt.xlabel("SONGS PLAYED")
plt.ylabel("LISTENING TIME")
plt.legend()
plt.show()
```

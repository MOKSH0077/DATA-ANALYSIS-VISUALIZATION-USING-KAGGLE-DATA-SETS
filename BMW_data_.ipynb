# BMW M5 SERIES 2022 KAGGLE DATASET



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bmw_sales=pd.read_csv("bmw.csv")
print(bmw_sales)
```

```python
bmw_m5=(bmw_sales['Model']=="5 Series") & (bmw_sales['Year']==2022) & (bmw_sales['Fuel_Type']=='Petrol')&(bmw_sales['Transmission']=='Manual')&(bmw_sales['Sales_Volume']<5000)
five_series=bmw_sales[bmw_m5]
print(five_series)
```

```python
plt.figure(figsize=(8,8))
sns.lineplot(x=five_series['Sales_Volume'],y=five_series['Price_USD'],hue=five_series['Region'])

plt.title("SALES VS PRICE FOR BMW M5 SERIES")
plt.xlabel("SALES VOLUME")
plt.ylabel("PRICE IN $")
plt.legend()
plt.style.use('dark_background')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.barplot(x=five_series['Sales_Volume'],y=five_series['Price_USD'],hue=five_series['Color'])
plt.title("SALES VS PRICE FOR BMW M5 SERIES")
plt.xlabel("SALES VOLUME")
plt.ylabel("PRICE IN $")
plt.legend()
plt.style.use('dark_background')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x=five_series["Engine_Size_L"],y=five_series['Mileage_KM'],hue=five_series['Color'])
plt.title(" ENGINE SIZE VS MILEAGE FOR BMW M5 SERIES")
plt.xlabel("ENGINE SIZE (L)")
plt.ylabel("MILEAGE")
plt.legend()
plt.style.use('dark_background')
plt.show()
```

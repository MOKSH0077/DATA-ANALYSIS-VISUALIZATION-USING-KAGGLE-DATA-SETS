```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sleep_data=pd.read_csv("sleep and health.csv")
print(sleep_data)
```

 # FOR MALE , AGE < 30

```python
sleep_data_male=(sleep_data['Gender']=='Male')&(sleep_data['Age']<30)
sleep_data_young_male=sleep_data[sleep_data_male]
print(sleep_data_young_male)
```

```python
plt.figure(figsize=(6,6))
plt.style.use('dark_background')
plt.title("DAILY STEPS VS Quality Of Sleep")
sns.scatterplot(x=sleep_data_young_male['Daily Steps'],y=sleep_data_young_male['Quality of Sleep'],hue=sleep_data_young_male['Occupation'],color='red')
plt.xlabel('Daily Steps')
plt.ylabel('Sleep Quality')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.style.use('dark_background')
plt.title("STRESS LEVEL VS DAILY STEPS")
sns.barplot(x=sleep_data_young_male['Stress Level'],y=sleep_data_young_male['Daily Steps'],hue=sleep_data_young_male['Occupation'])
plt.xlabel('Stress')
plt.ylabel('Daily Steps')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.style.use('dark_background')
plt.title("QUALITY OF SLEEP VS SLEEP DURATION")
sns.boxplot(x=sleep_data_young_male['Quality of Sleep'],y=sleep_data_young_male['Sleep Duration'],hue=sleep_data_young_male['Occupation'])
plt.xlabel('Quality of Sleep')
plt.ylabel('Sleep Duration')
plt.legend()
plt.show()
```
